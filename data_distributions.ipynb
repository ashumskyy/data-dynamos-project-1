```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
# Read in csv file
final_combined_variables_world_happiness_df = pd.read_csv("final_combined_variables_world_happiness.csv")
final_combined_variables_world_happiness_df.head()
```

```python
# Plot histogram for Life Expectancy
plt.hist(final_combined_variables_world_happiness_df['Life Expectancy'],color="royalblue")
plt.xlabel('Life Expectancy')
plt.ylabel('Number of Countries')

plt.savefig('output/life_expectancy_dist_2.png', dpi=800)
plt.show()
```
