```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
summary = pd.read_csv("output/final_combined_variables_world_happiness.csv")
summary_df = pd.DataFrame(summary)
summary_df.head()
```

```python
summary_decending = summary_df.sort_values(["Happiness Score"],ascending = [False])
summary_decending.reset_index(drop=True)
```

```python
summary_decending.info()
```

```python
#top_20_df = summary_decending[(summary_decending['Happiness Score']>=6.687)]
top_15_df = summary_decending.iloc[0:15]
top_15_df.reset_index(drop=True)
```

```python


top_15_df.plot(x="Country Name", y="Happiness Score", ylabel="Happiness Score", kind="bar",color='limegreen',
               align="center",figsize=(10,5))
plt.xticks(rotation=45, ha='right')
plt.savefig("output/top_15_happy_countries.png")
plt.show()
```

```python
bottom_15_df = summary_decending[(summary_decending['Happiness Score']<4.394)]
bottom_15_df                                 
```

```python
bottom_15_df.plot(x="Country Name", y="Happiness Score", ylabel="Happiness Score", kind="bar",color='darkgreen',
                  align="center",figsize=(10,5))
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 8)
plt.savefig("output/bottom_15_happy_countries.png")
plt.show()
```

```python
summary_decending_suicide = summary_df.sort_values(["Suicide Rate"],ascending = [False])
summary_decending_suicide.reset_index(drop=True)
```

```python
# top_20_suicide = summary_decending_suicide[(summary_decending_suicide['Happiness Score']<4.543)]
highest_15_suicide = summary_decending_suicide.iloc[0:15]
highest_15_suicide.reset_index(drop=True)
```

```python
summary_ascending_suicide = summary_df.sort_values(["Suicide Rate"])
summary_ascending_suicide.reset_index(drop=True)
```

```python
# bottom_20_suicide = summary_decending_suicide[(summary_decending_suicide['Happiness Score']<4.543)]
lowest_15_suicide = summary_ascending_suicide.iloc[0:15]
lowest_15_suicide.reset_index(drop=True)
```

```python

lowest_15_suicide.plot(x="Country Name", y="Suicide Rate", ylabel="Suicide Rate (per 100,000 pop)", 
                       kind="bar",color='navy',align="center",figsize=(10,5))
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 10)
plt.savefig("output/lowest_15_suicide_countries.png")
plt.show()
```

```python
highest_15_suicide.plot(x="Country Name", y="Suicide Rate", ylabel="Suicide Rate (per 100,000 pop)", 
                        kind="bar",color='royalblue',align="center",figsize=(10,5))

plt.xticks(rotation=45, ha='right')
plt.figure(figsize=(20,6))
# plt.ylim(0, 8)
plt.savefig("output/highest_15_suicide_countries.png")
plt.show()
```
