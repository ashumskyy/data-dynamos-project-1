### Create Final Summary Table with World Happiness Score and the 6 Socio/Economic Variables

```python
# Importing dependencies
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in CSV files
combined_mean_life_exp_suicide_unemploy = pd.read_csv("output/combined_mean_life_exp_suicide_unemploy.csv")
oleksiis_three_merged = pd.read_csv("output/oleksiis_three_merged_df.csv")
world_happiness_score_2022_clean_names = pd.read_csv("Resources/world_happiness_score_2022_clean_names.csv")
```

```python
# Convert to dataframes
combined_mean_life_exp_suicide_unemploy_df = pd.DataFrame(combined_mean_life_exp_suicide_unemploy)
oleksiis_three_merged_df = pd.DataFrame(oleksiis_three_merged)
world_happiness_score_2022_clean_names_df = pd.DataFrame(world_happiness_score_2022_clean_names)
```

```python
# Preview world_happiness dataframe
world_happiness_score_2022_clean_names_df
```

```python
# Delete unwanted columns
world_happiness_score_2022_clean_names_drop_df = world_happiness_score_2022_clean_names_df.drop(columns=['RANK', 'Whisker-high',
                                                                                 'Whisker-low',
                                                                                 'Dystopia (1.83) + residual',
                                                                                 'Explained by: GDP per capita',
                                                                                 'Explained by: Social support',
                                                                                 'Explained by: Healthy life expectancy',
                                                                                 'Explained by: Freedom to make life choices',
                                                                                 'Explained by: Generosity',
                                                                                 'Explained by: Perceptions of corruption'
                                                                                ])

# Rename "Country" column to "Country Name"
world_happiness_score_2022_clean_names_drop_rename_df = world_happiness_score_2022_clean_names_drop_df.rename(columns={'Country' : 'Country Name'})

world_happiness_score_2022_clean_names_drop_rename_df
```

```python
# Merge world_happiness with combined_mean_life_exp_suicide_unemploy_df dataframe

combined_mean_world_happiness_df = pd.merge(world_happiness_score_2022_clean_names_drop_rename_df, 
                                             combined_mean_life_exp_suicide_unemploy_df, 
                                             how="outer", on=["Country Name", "Country Name"])
```

```python
# Merge oleksiis_three_merged_df with combined_mean_world_happiness_d dataframe
combined_mean_world_happiness_2_df = pd.merge(combined_mean_world_happiness_df, 
                                             oleksiis_three_merged_df, 
                                             how="outer", on=["Country Name", "Country Name"])
```

```python
# Preview the dataframe
combined_mean_world_happiness_2_df.head()
```

```python
# Delete unwanted columns
combined_mean_world_happiness_2_drop_df = combined_mean_world_happiness_2_df.drop(columns=['Unnamed: 0_x', 
                                                                                           'Unnamed: 0_y',
                                                                                           'Country Code_y',
                                                                                ])

combined_mean_world_happiness_2_drop_df.head()
```

```python
# Rename "Country Code-x" to "Country Code"
combined_mean_world_happiness_3_df = combined_mean_world_happiness_2_drop_df.rename(columns={
    'Country Code_x' : 'Country Code'})

combined_mean_world_happiness_3_df
```

```python
# # Replace all empty cells  with NaN
combined_mean_world_happiness_4_df = combined_mean_world_happiness_3_df.replace('', np.nan)


# Drop all NaN values
combined_mean_world_happiness_dropNaN_df = combined_mean_world_happiness_4_df.dropna(axis=0)

combined_mean_world_happiness_dropNaN_df
```

```python
# Move Country Code column after Country Name
combined_mean_world_happiness_movecol_df = combined_mean_world_happiness_dropNaN_df[['Country Name',
                                                                                    'Country Code',
                                                                                    'Happiness score',
                                                                                    'LifeExpMean',
                                                                                    'SuicideMean',
                                                                                    'UnempMean',
                                                                                    'NetIncomeMean',
                                                                                    'GDP',
                                                                                    'Inflation']]

combined_mean_world_happiness_movecol_df

```

```python
# Rename Columns
final_combined_variables_world_happiness_df = combined_mean_world_happiness_movecol_df.rename(columns={
    'Happiness score' : 'Happiness Score',
    'LifeExpMean' : 'Life Expectancy',
    'SuicideMean' : 'Suicide Rate',
    'UnempMean' : 'Unemployment Rate',
    'NetIncomeMean' : 'Net National Income per capita',
    'GdpMean' : 'GDP per capita',
    'InflMean' : 'Consumer Price Inflation'
})

final_combined_variables_world_happiness_df

```

```python
# Write final summary dataframe to csv file
final_combined_variables_world_happiness_df.to_csv('output/final_combined_variables_world_happiness.csv', index=True)
```
