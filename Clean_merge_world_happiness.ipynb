```python
# Import dependencies
# Importing dependencies
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in csv files ** MAKE SURE THIS IS CORRECT WHEN COMMITTED TO MAIN BRANCH **
combined_life_exp_suicide_unemploy = pd.read_csv("Resources/combined_mean_life_exp_suicide_unemploy.csv")
world_happiness_score_revised = pd.read_csv("Resources/world_happiness_score_2022_revised_names.csv")
```

```python
# Convert data to DataFrames
combined_life_exp_suicide_unemploy_df = pd.DataFrame(combined_life_exp_suicide_unemploy)
world_happiness_score_revised_df = pd.DataFrame(world_happiness_score_revised)
```

```python
# Preview dataframes
combined_life_exp_suicide_unemploy_df
```

```python
# Preview dataframes
world_happiness_score_revised_df
```

```python
# View the data types
world_happiness_score_revised_df.info()
```

```python
# Drop unwanted columns

world_happiness_score_dropped_df = world_happiness_score_revised_df.drop(columns=['RANK','Whisker-high',
                                                                          'Whisker-low','Dystopia (1.83) + residual',
                                                                          'Explained by: GDP per capita',
                                                                          'Explained by: Social support',
                                                                          'Explained by: Healthy life expectancy',
                                                                          'Explained by: Freedom to make life choices',
                                                                          'Explained by: Generosity',
                                                                          'Explained by: Perceptions of corruption'           
                                                                         ])

world_happiness_score_dropped_df
```

```python
# Rename "Country" column to "Country Name"
world_happiness_score_renamed_df = world_happiness_score_dropped_df.rename(columns={'Country' : 'Country Name'})
world_happiness_score_renamed_df
```

```python
# use OUTER to merge 2 dataframes keeping all rows even if not matched;  changed "HOW':left"  to "HOW:OUTER"
# school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Merge World Happiness to Combined Life Exp/Suicide/Unemploy df 
rev_combined_life_exp_suic_unemp_happy_df = pd.merge(world_happiness_score_renamed_df,
                                                 combined_life_exp_suicide_unemploy_df, 
                         how="outer", on=["Country Name", "Country Name"])

rev_combined_life_exp_suic_unemp_happy_df
```

```python
# Write data frame to CSV file to compare Country Names
rev_combined_life_exp_suic_unemp_happy_df.to_csv('Resources/rev_combined_life_exp_suic_unemp_happy.csv', index=True)
```

```python
# Replace all empty values with NaN
combined_life_exp_suic_unemp_happy_dropped_df = rev_combined_life_exp_suic_unemp_happy_df.replace(' ', np.nan)

# Drop all NaN values
combined_life_exp_suic_unemp_happy_clean_df = combined_life_exp_suic_unemp_happy_dropped_df.dropna(axis=0)
```

```python
combined_life_exp_suic_unemp_happy_clean_df.to_csv('Resources/combined_life_exp_suic_unemp_happy_clean.csv', index=True)
```
